```python
import pandas as pd
import numpy as np
```

```python
# read train set
train = pd.read_csv('train.csv')
print(train.info())
print(train.head(5))
```

```python
train.describe()
```

```python
print(pd.isnull(train).sum())
```

```python
nan = train.Cabin[0]
def objToNum(df, attr):
    names = list(df[attr].value_counts().index)
    if pd.isnull(df).sum()[attr]:
        names.append(nan)
    nums = range(len(names))
    dic = dict(zip(names, nums))
    dic[nan] = nan
    df[attr] = df[attr].map(lambda x: dic[x])
```

```python
# data preprocessing
from sklearn.model_selection import train_test_split
# 5 attributes to handle, Name, Sex, Ticket, Cabin, Embarked
train.drop(['Name', 'Ticket', 'Cabin'],axis=1,inplace=True)
objToNum(train, 'Sex')
objToNum(train, 'Embarked')
```

```python
# remove null samples
train.dropna(axis=0, how='any', inplace=True)
```

```python
x=train.drop('Survived',axis=1,inplace=False)
y=train[['Survived']]
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.3,random_state=0)
```

```python
# random forest
from sklearn.ensemble import RandomForestClassifier
rf = RandomForestClassifier()
rf.fit(x_train, y_train)
rf.score(x_test, y_test)
```

```python
# Decision Tree
from sklearn.tree import DecisionTreeClassifier
df = DecisionTreeClassifier()
df.fit(x_train, y_train)
df.score(x_test, y_test)
```

```python
from sklearn.neighbors import KNeighborsClassifier
neigh = KNeighborsClassifier(n_neighbors=20)
neigh.fit(x_train, y_train)
neigh.score(x_test, y_test)
```

```python
from sklearn.linear_model import LogisticRegression
logmodel=LogisticRegression()
logmodel.fit(x_train,y_train)
logmodel.score(x_test, y_test)
```

```python
from sklearn.svm import LinearSVC
svc = LinearSVC(penalty='l2', loss='squared_hinge', random_state=0, max_iter=10e4)
svc.fit(x_train,y_train)
svc.score(x_test, y_test)
```

```python
test = pd.read_csv('test.csv')
# 5 attributes to handle, Name, Sex, Ticket, Cabin, Embarked
test.drop(['Name', 'Ticket', 'Cabin'],axis=1,inplace=True)
objToNum(test, 'Sex')
objToNum(test, 'Embarked')
```

```python
test.dropna(axis=0, how='any', inplace=True)
```
